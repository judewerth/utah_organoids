#### Understanding the Pipeline Structure
The Utah Organoids pipeline is structured into four main schemas:
1. `Lineage` – Tracks the biological history and derivation of organoids.
2. `Culture` – Stores experiment conditions and metadata for cultured organoids.
3. `Probe` – Defines electrophysiology probe configurations and channel mappings.
4. `Ephys` – Manages electrophysiology session data, spike sorting, and analysis.
